# Visualizing the dataset

```python
fig, ax = plt.subplots(nrows = 2, ncols = 3)
fig.tight_layout()
fig.set_figheight(15)
fig.set_figwidth(15)

```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 3)
happiness_score = df['Happiness Score']
GDP = df['GDP per capita']
Social_Support = df['Social support']
Healthy_life_expectancy =df['Healthy life expectancy']
Freedom = df['Freedom to make life choices']
Generosity = df['Generosity']
Perceptions_of_corruption = df['Perceptions of corruption']

fig.tight_layout()
fig.set_figheight(15)
fig.set_figwidth(15)

ax[0,0].scatter(happiness_score, GDP)
ax[0,1].scatter(happiness_score, Social_Support)
ax[0,2].scatter(happiness_score, Healthy_life_expectancy)
ax[1,0].scatter(happiness_score, Freedom)
ax[1,1].scatter(happiness_score, Generosity)
ax[1,2].scatter(happiness_score, Perceptions_of_corruption)



ax[0,0].set_xlabel('Happiness Score')
ax[0,1].set_xlabel('Happiness Score')
ax[0,2].set_xlabel('Happiness Score')
ax[1,0].set_xlabel('Happiness Score')
ax[1,1].set_xlabel('Happiness Score')
ax[1,2].set_xlabel('Happiness Score')


ax[0,0].set_ylabel('GDP')
ax[0,1].set_ylabel('Social Support')
ax[0,2].set_ylabel('Healthy life expectancy')
ax[1,0].set_ylabel('Freedom')
ax[1,1].set_ylabel('Generosity')
ax[1,2].set_ylabel('Perceptions of corruption')

ax[0,0].title.set_text('GDP vs Happiness Score')
ax[0,1].title.set_text('Social Support vs Happiness Score')
ax[0,2].title.set_text('Healthy life expectancy vs Happiness Score')
ax[1,0].title.set_text('Freedom vs Happiness Score')
ax[1,1].title.set_text('Generosity vs Happiness Score')
ax[1,2].title.set_text('Perceptions of corruption vs Happiness Score');


```
